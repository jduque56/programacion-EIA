```python
import matplotlib.pyplot as plt
import csv

#inicializo las variables y listas necesarias para crear desarrollar mi codigo
latonR=[]
ticket=[]
label=[]
x=[]
y=[]
aux=0.0

#leo mi archivo csv y le doy los valores necesarios a mis listas
with open('archivos_csv/laton_rojo.csv','r') as csvfile:
    csvreader=csv.reader(csvfile,delimiter=';')
    for row in csvreader:
        latonR.append(row)
        x.append(row[2])
        y.append(row[1])
        ticket.append(row[0])
label=latonR[0]

#remuevo de mis listas valores que no me interesan
ticket.remove('espesor')
x.remove('Dureza_Promedio')
y.remove('%TF')
label.remove('espesor')

#convierto los valor de string a float
for idx in range(len(x)):
    aux=float(x[idx])
    x[idx]=aux
for idx in range(len(y)):
    aux=float(y[idx])
    y[idx]=aux

#grafico
plt.plot(y,x)
plt.xticks(y,)
plt.yticks(x)
plt.xlabel(label[0])
plt.ylabel(label[1])
plt.show()
```
