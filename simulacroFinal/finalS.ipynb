```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed()
bolasColores=np.array(['R','R','R','A','A','V'])
rojo=0
azul=0
verde=0
for idx in range(0,100000,1):
    numero=np.random.randint(6)
    if bolasColores[numero]=='R':
        rojo+=1
    elif bolasColores[numero]=='V':
        verde+=1
    elif bolasColores[numero]=='A':
        azul+=1

vecesRepetidas=np.array([azul,rojo,verde])
color=np.array(['Azul','Rojo','Verde'])
print(vecesRepetidas)
plt.bar(x=color,height=vecesRepetidas)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(50)
T = 30*np.random.random(1000) + 5
RH = 100*np.random.random(1000)

fig, ax1 = plt.subplots()

ax1.plot(T, color='red')
ax1.set_ylabel('Temperatura (°C)', color='red')
#ax1.tick_params('y', colors='red')

ax2 = ax1.twinx()
ax2.plot(RH, color='blue')
ax2.set_ylabel('Humedad Relativa (%)', color='blue')
#ax2.tick_params('y', colors='blue')

# Agregar líneas horizontales para HR 80% y temperatura 20 °C
ax2.axhline(y=49.66, color='black', linestyle='--')
ax2.axhline(y=80, color='black', linestyle='--')

plt.title('Datos de Temperatura y Humedad Relativa')
plt.show()
```
