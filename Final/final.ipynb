#### Escriba un programa que lea el contenido del archivo AutomobileData.csv y calcule los momentos para las variables horsepower, average_mileage y price. Puede validar sus cálculos usando la función moment de la librería SciPy.


```python
Automobile = pd.read_csv('archivos/AutomobileData.csv',sep=';')
datosUtilizados = pd.DataFrame({'horsepower':Automobile.horsepower,'avarage':Automobile.average_mileage,'price':Automobile.price})
n = 61
momentosHP = pd.DataFrame()
momentosAM = pd.DataFrame()
momentosP = pd.DataFrame()
datosCentrados = datosUtilizados - datosUtilizados.mean()
for idx in range(0,5,1):
    HP = ((datosCentrados.horsepower)**idx).sum()
    momentosHP[str(idx)] = [HP/n]

    AM = ((datosCentrados.avarage)**idx).sum()
    momentosAM[str(idx)] = [AM/n]

    P = ((datosCentrados.price)**idx).sum()
    momentosP[str(idx)] = [P/n]
```

#### El archivo LoremIpsum.txt contiene cinco párrafos del texto de relleno Lorem Ipsum. Desarrolle un programa en Python que lea el contenido del archivo, cuente la cantidad de veces que aparecen las cadenas qu, um, or, si, su, ul, er, ar, cu, en (case insensitive) y muestre el resultado usando diagrama de barras, con los datos organizados en orden descendente.

```python
# Leer el contenido del archivo de texto
with open('archivos/LoremIpsum.txt', 'r') as file:
    content = file.read()

# Convertir el contenido a minúsculas
content = content.lower()

# Cadenas a buscar
cadenas = ['qu', 'um', 'or', 'si', 'su', 'ul', 'er', 'ar', 'cu', 'en']

# Contar la cantidad de veces que aparecen las cadenas
resultados = {}
for cadena in cadenas:
    count = content.count(cadena)
    resultados[cadena] = count

# Crear un DataFrame de pandas con los resultados
df = pd.DataFrame(resultados.items(), columns=['Cadena', 'Cantidad'])

# Ordenar el DataFrame en orden descendente
df = df.sort_values('Cantidad', ascending=False)

# Mostrar el diagrama de barras
plt.figure(figsize=(10, 6))
plt.bar(df['Cadena'], df['Cantidad'])
plt.xlabel('Cadena')
plt.ylabel('Cantidad')
plt.title('Frecuencia de cadenas en Lorem Ipsum')
plt.xticks(rotation=45)
plt.show()
```

#### Ahora trabajará con el archivo N170OnewayANOVA.csv. Cargue la información en un DataFrame y posteriormente:

    (b) (5 puntos) cree un scatter plot de la Amplitud vs. Latencia, usando un color diferente para cada

```python
NEW = pd.read_csv('archivos/N170OnewayANOVA.csv',sep=';')

# Crear visualización de boxplot para la columna Lat_ms agrupada por Stimulus
plt.figure(figsize=(10, 6))
NEW.boxplot(column='Lat_ms', by='Stimulus')
plt.xlabel('Stimulus')
plt.ylabel('Latencia (ms)')
plt.title('Boxplot de Latencia por Tipo de Estímulo')
plt.xticks(rotation=45)
plt.show()

# Crear visualización de boxplot para la columna Amp_uV agrupada por Stimulus
plt.figure(figsize=(10, 6))
NEW.boxplot(column='Amp_uV', by='Stimulus')
plt.xlabel('Stimulus')
plt.ylabel('Amplitud (µV)')
plt.title('Boxplot de Amplitud por Tipo de Estímulo')
plt.xticks(rotation=45)
plt.show()


#Scatter
# Crea un diccionario de colores para los tipos de estímulo
colors = {'Famous': 'red', 'Unfamiliar': 'blue', 'Scrambled': 'green'}

# Crea el scatter plot con colores diferentes para cada tipo de estímulo
plt.figure(figsize=(10, 6))
for stimulus, color in colors.items():
    subset = NEW[NEW['Stimulus'] == stimulus]
    plt.scatter(subset['Amp_uV'], subset['Lat_ms'], color=color, label=stimulus)

# Personaliza el gráfico
plt.xlabel('Amplitud (µV)')
plt.ylabel('Latencia (ms)')
plt.title('Scatter plot de Amplitud vs. Latencia')
plt.legend()

# Muestra el gráfico
plt.show()




```
